```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_html("https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C",
                   attrs={"class": "wikitable collapsible collapsed", "style": "text-align: right"}, match="Регіон",
                   decimal=',', thousands='.')
df = df[0]
df
```

```python
#Вивести перші рядки таблиці за допомогою методу head
df.head(5)
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df.shape
```

```python
#Замініть у таблиці значення "—" на значення NaN
df = df.replace({"—": np.nan})
df
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
#Замініть типи нечислових колонок на числові
df['2014'] = pd.to_numeric(df['2014'])
df['2019'] = pd.to_numeric(df['2019'])
df.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum()/df.shape[0]
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
df.drop(axis=0, index=27, inplace=True)
df
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df = df.fillna({"1950": df["1950"].mean(), "1960": df["1960"].mean(), "1970": df["1970"].mean(),
                "1990": df["1990"].mean(), "2000": df["2000"].mean(), "2012": df["2012"].mean(),
                "2014": df["2014"].mean(), "2019": df["2019"].mean()})
df
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
df[df["2019"] > df["2019"].mean()][["Регіон", "2019"]].sort_values("2019")
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
df.sort_values(["2014"], ascending=False).head(1)[["Регіон", "2014"]]
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
fig, ax = plt.subplots(figsize=(16, 10))
df.plot(x="Регіон", y="2019", kind="bar",
        fontsize=14, color='g', ax=ax, grid=True)
plt.title("Діаграма народжуваності по регіонах у 2019 році", fontdict = {'fontsize' : 18})
ax.set_xlabel("Регіон", size=16, color="midnightblue")
ax.set_ylabel("Коефіцієнт народжуваності",size=16, color="midnightblue")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
date = pd.date_range(start="1950", freq="10Y", periods=8)
k_per_year = [20.4, 18.9, 15.6, 12.3, 7.3, 12.2, 12.1, 8.0]
plt.plot(date, k_per_year, label='Коефіцієнт народжуваності', linestyle="--", marker="^", color="green")
plt.xlabel('Дата', fontsize=15, color='midnightblue')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='midnightblue')
plt.title('Коефіцієнт народжуваності в Київській обл', fontsize=15)
plt.text(date[4], 7.5, 'Мінімальний коефіцієнт', color="red", fontsize=12)
plt.legend()
plt.show()
```

```python
sort_df = df.sort_values(['2019'], ascending=False)
fig, ax = plt.subplots(figsize=(16, 14))
sns.pointplot(x="2019", y="Регіон", data=sort_df)
plt.title('Коефіцієнт народжуваності в 2019 по обл', fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
sns.lineplot(data=df, marker="D")
plt.xlabel('Індекс області', fontsize=15, color='midnightblue')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='midnightblue')
plt.title('Коефіцієнт народжуваності в період з 1950-го по 2019-й рік', fontsize=15)
plt.show()
```
